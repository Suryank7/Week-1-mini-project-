```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
import seaborn as sns
```

```python
df = pd.read_csv("athlete_events.csv")
```

```python
noc = pd.read_csv("noc_regions.csv")
```

```python
df = df.merge(noc, on= "NOC", how= "left")
```

```python
df.head()
```

```python
medals = df[df['Medal'].notnull()]
print("Total medals recorded:", medals.shape[0])
```

```python
medals.head()
```

```python
country_medals = medals.groupby("region")["Medal"].count().sort_values(ascending = False).head(10)
plt.figure(figsize = (10,6))
sns.barplot(x = country_medals.values, y= country_medals.index, palette = "viridis")
plt.title("Top 10 countries by Total Olympic Medals")
plt.xlabel("Medala Won")
plt.ylabel("Country")
plt.show()
```

```python
yearly_medals = medals.groupby(["Year","region"])["Medal"].count().reset_index()
plt.figure(figsize=(12,6))
sns.lineplot(data = yearly_medals[yearly_medals['region'].isin(["USA","China","Russia","India","UK"])], x="Year", y="Medal", hue="region", marker="o")
plt.title("Olympic Medal Trends Over Time (Selected Countries)")
plt.xlabel("Year")
plt.ylabel("Medals Won")
plt.show()
```

```python
medal_split = medals.groupby(["region","Medal"])["ID"].count().unstack().fillna(0).sort_values("Gold", ascending= False).head(5)
medal_split.plot(kind="bar", stacked=True, figsize=(10,6), colormap="Set2")
plt.title("Top 5 Countries - Medal Type Distribution")
plt.xlabel("Country")
plt.ylabel("Count")
plt.show()
```
